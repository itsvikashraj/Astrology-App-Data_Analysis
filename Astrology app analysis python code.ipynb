```python
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://drive.google.com/uc?id=1whoSp5NEdmj5AdXcGj7eTSrYyf6k8O4s")
```

```python
columns_to_drop = [
    "callSeconds", "complementaryCallOrganiser", "Remedies", "cardPosition", "RecordingUrl", "disconnectedBy"
]
df = df.drop(columns=columns_to_drop)
```

```python
df['chatStatus'].fillna('Unknown', inplace=True)
```

```python
# Fill categorical columns with "Unknown"
categorical_columns = ['chatStatus', 'callStatus', 'source', 'refundStatus']
for col in categorical_columns:
    df[col].fillna('Unknown', inplace=True)
```

```python

missing_percentage = df.isnull().mean() * 100
missing_percentage
```

```python
# List of numerical columns
numerical_columns = ['chatSeconds', 'timeDuration', 'amount', 'astrologerOnCallDuration']

for col in numerical_columns:
    # Check if the column is skewed
    if df[col].skew() < 0.5 and df[col].skew() > -0.5:  # Approximately symmetric
        df[col].fillna(df[col].mean(), inplace=True)  # Use mean
    else:
        df[col].fillna(df[col].median(), inplace=True)  # Use median
```

```python
# Calculate the median percentage of missing values across columns
missing_percentage = df.isnull().mean() * 100
missing_percentage
```

```python
# Drop columns with high missing values
columns_to_drop = [
    "callChannel", "callIvrType", "CallSid",
    "userOnCallDuration", "feedback"
]
df_cleaned = df.drop(columns=columns_to_drop)

categorical_columns = ["statementEntryId", "astrologerCallStatus", "hideHistory", "offer", "userCallStatus"]
for col in categorical_columns:
    df_cleaned[col] = df_cleaned[col].fillna("Unknown")

datetime_columns = ["chatStartTime", "chatEndTime"]
for col in datetime_columns:
    df_cleaned[col] = pd.to_datetime(df_cleaned[col], errors='coerce')
    df_cleaned[col].fillna(df_cleaned[col].min(), inplace=True)  # Fill with earliest datetime

numerical_columns = ["astrologersEarnings", "netAmount"]
for col in numerical_columns:
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
    df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)

print(df_cleaned.isnull().sum())
```

```python
missing_percentage = df_cleaned.isnull().mean() * 100
missing_percentage
```

```python
df_cleaned.drop('region', axis=1, inplace=True)
```
